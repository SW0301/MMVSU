```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

h = 0.01

def dydt(y, t):
    k=0.42
    return k * y

def ode(dydt,y0,t,n):
  y=[y0]
  for i in range(n-1):
    ti=t[i]
    yi=y[i]
    yi1=yi + h * dydt(yi,ti)
    y.append(yi1)
  return y


t = np.arange(0, 10+h, h) 
y = ode(dydt, y0=1, t=t, n=len(t-1))
#print(y)
t1 = np.linspace(0,10,100)
solve = odeint(dydt, y0=1, t=t1)
solve = solve.flatten()
#print(solve)
plt.plot(t1,solve, label = r"odeint")

plt.plot(t, y, label = r"euler")
plt.grid(True)
plt.legend()
plt.show()
```
