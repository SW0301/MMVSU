```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import pandas as pd

url = 'https://raw.githubusercontent.com/SW0301/MMVSU/main/lab2.csv'
df = pd.read_csv(url)

years = np.array(df['Year'])
population = np.array(df['Population'])


def k(x):
    return x/3060000 
def dxdt(x, t):
    return k(x) * x

t = np.linspace(1811, 2021, 100) 

solve = odeint(dxdt, y0=14400, t=t)
solve = solve.flatten()

plt.plot(t, solve+10000)
plt.plot(years, population)
plt.show()
```
